```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset (avoids file error)
data = {
    'Airline': ['IndiGo','Air India','SpiceJet','IndiGo','Air India',
                'SpiceJet','IndiGo','Air India','SpiceJet','IndiGo'],
    'Origin': ['Chennai','Delhi','Mumbai','Chennai','Delhi',
               'Mumbai','Chennai','Delhi','Mumbai','Chennai'],
    'Destination': ['Delhi','Mumbai','Chennai','Mumbai','Chennai',
                    'Delhi','Delhi','Mumbai','Chennai','Mumbai'],
    'Departure Delay': [10, -5, 20, 0, 15, 30, -2, 5, 25, 8]
}

df = pd.DataFrame(data)

# Save CSV
df.to_csv('flight_data.csv', index=False)

# Load dataset
df = pd.read_csv('flight_data.csv')

# a) Summary statistics for departure delays
print("Summary Statistics for Departure Delays:\n")
print(df['Departure Delay'].describe())

# b) Average delay for each airline
avg_delay = df.groupby('Airline')['Departure Delay'].mean()

print("\nAverage Delay for Each Airline:\n")
print(avg_delay)

# c) Most common origin and destination
most_common_origin = df['Origin'].mode()[0]
most_common_destination = df['Destination'].mode()[0]

print("\nMost Common Origin:", most_common_origin)
print("Most Common Destination:", most_common_destination)

# d) Box plot for departure delays
plt.figure()
plt.boxplot(df['Departure Delay'])

plt.title('Distribution of Departure Delays')
plt.ylabel('Delay (minutes)')

plt.show()
```
