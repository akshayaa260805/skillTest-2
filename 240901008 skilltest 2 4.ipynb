```python
import math

# ----------- Euclidean Distance in 2D -----------

# Input points
x1, y1 = 2, 3
x2, y2 = 6, 7

# Formula: √((x2-x1)^2 + (y2-y1)^2)
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Euclidean Distance:", distance)


# ----------- Dot Product of Two Matrices -----------

# Matrix A (2x2)
A = [
    [1, 2],
    [3, 4]
]

# Matrix B (2x2)
B = [
    [5, 6],
    [7, 8]
]

# Result matrix initialized with zeros
result = [
    [0, 0],
    [0, 0]
]

# Matrix multiplication (dot product)
for i in range(len(A)):          # rows of A
    for j in range(len(B[0])):   # columns of B
        for k in range(len(B)):  # rows of B
            result[i][j] += A[i][k] * B[k][j]

print("\nDot Product (Matrix Multiplication Result):")
for row in result:
    print(row)
```
