```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------- Step 1: Create Time Series Data -----------

# Generate date range for one year
dates = pd.date_range(start='2025-01-01', periods=365, freq='D')

# Generate random temperature data using NumPy
np.random.seed(0)
temperatures = np.random.normal(loc=25, scale=5, size=365)

# Create DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures
})

# Set Date as index
df.set_index('Date', inplace=True)

print("First 5 rows of dataset:\n")
print(df.head())


# ----------- Step 2: Basic Analysis -----------

# Overall statistics
print("\nSummary Statistics:\n")
print(df.describe())

# Monthly average temperature
monthly_avg = df.resample('M').mean()

print("\nMonthly Average Temperature:\n")
print(monthly_avg)


# ----------- Step 3: Visualization -----------

# Line plot of daily temperatures
plt.figure()
plt.plot(df.index, df['Temperature'])
plt.title('Daily Temperature Over a Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()

# Bar plot for monthly averages
plt.figure()
monthly_avg.plot(kind='bar')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```
